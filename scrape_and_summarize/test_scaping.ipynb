```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.hindustantimes.com/'
headers = {
    "User-Agent": "Mozilla/5.0",
}
```

```python
page = requests.get(url, headers=headers)
```

```python
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
all_links = soup.find_all('a', class_='storyLink articleClick')
```

```python
len(all_links)
```

```python
first_five = all_links[:5]
```

```python
for link in first_five:
    print(link['href'])
```

```python
news_links = url + first_five[1]['href'][1:]

news_links
```

```python
specific_page = requests.get(news_links, headers=headers)

soup2 = BeautifulSoup(specific_page.text, 'html.parser')
```

```python
soup2.prettify()
```

```python
heading = soup2.find('h1', class_='hdg1').text.strip()
heading
```

```python
headings = soup2.find_all('h1', class_='hdg1')
len(headings)
```

```python
paragraphs = soup2.find_all('p')
len(paragraphs)
```

```python
news = ""
for paragraph in paragraphs[1:]:
    news += paragraph.text.strip() + "\n"
```

```python
print(news)
```
